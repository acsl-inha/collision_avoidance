```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sc = 8000
cc = -2000
cos_start = 10
x_c = np.arange(-100, 100, 0.1)
y_c = (sc - cc) / 2 * np.sin(np.pi / (100 - cos_start / 2) * (x_c - (100 + cos_start) / 2)) \
        + (sc + cc) / 2
plt.figure(figsize = (15, 15))
plt.plot(x_c, y_c)
plt.grid()
```

```python
sc = 8000
cc = -2000
cos_start = 80
```

```python
x_p = np.arange(100, 200, 0.1)
x_c = np.arange(cos_start, 100, 0.1)
x_n = np.arange(-100, cos_start, 0.1)
```

```python
y_p = np.full((len(x_p)), sc)
y_c = (sc - cc) / 2 * np.sin(np.pi / (100 - cos_start) * (x_c - (100 + cos_start) / 2)) \
        + (sc + cc) / 2
y_n = np.full((len(x_n)), cc)
```

```python
y_plot = np.concatenate([y_n, y_c ,y_p])
x_plot = np.concatenate([x_n, x_c, x_p])
```

```python
plt.figure(figsize = (15, 15))
plt.plot(x_plot, y_plot)
plt.grid()
# plt.xlim(90, 200)
```
