```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
rl_up = mpimg.imread("rl_up.png")
av_up = mpimg.imread("av_up.png")
im_up = mpimg.imread("im_up.png")
rl_down = mpimg.imread("rl_down.png")
av_down = mpimg.imread("av_down.png")
im_down = mpimg.imread("im_down.png")
```

```python
images = [av_up, im_up, rl_up]
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize = (18,6),gridspec_kw={'wspace':0, 'hspace':0},
                           squeeze=True)
for i, image in enumerate(images):
    axs[i].axis("off")
    axs[i].imshow(image)
axs[0].set_title("Domain knowledge algorithm")
axs[1].set_title("Supervised learning model")
axs[2].set_title("Reinforcement learning model")
fig.text(0.5, 0, 'Minimum distance', ha='center', fontsize=12)
fig.text(0, 0.5, 'Number of samples', va='center', rotation='vertical', fontsize=12)
plt.tight_layout()
plt.savefig("up_org.svg", format = 'svg', dpi=300)
```
