```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_p = np.arange(100,200,0.1)
x_n = np.arange(0,100,0.1)
```

```python
max_reward = 900000
y_p = max_reward * np.log((x_p - 90) / 10 * np.exp(1)) / (10 * (x_p - 90))
y_n = (10000 * 0.02 * (10000 * np.log(( - x_n + 110) / 10 * np.exp(1))/(10 * ( - x_n + 110)) - 10000 * 0.009) - 2000) + max_reward * 0.01
```

```python
y_plot = np.concatenate([y_n, y_p])
x_plot = np.concatenate([x_n, x_p])
```

```python
plt.figure(figsize = (15, 15))
plt.plot(x_plot, y_plot)
plt.grid()
```
